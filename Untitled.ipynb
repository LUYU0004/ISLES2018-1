```python
import os
from glob import glob
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

from scipy.ndimage.interpolation import rotate
% matplotlib inline
```

```python
image_list = glob("data/TRAINING/case*/*DWI*/*.nii")
label_list = glob("data/TRAINING/case*/*OT*/*.nii")
```

```python
label_path = label_list[0]

label_data = nib.load(label_path)
label = np.array(label_data.get_data())

image_path = image_list[0]

image_data = nib.load(image_path)
image = np.array(image_data.get_data())
```
