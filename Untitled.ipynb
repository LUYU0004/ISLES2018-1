```python
import os
from glob import glob
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
image_list = glob("training/*/pre/T1.nii.gz")
label_list = glob("training/*/segm.nii.gz")
```

```python
label_path = label_list[0]

label_data = nib.load(label_path)
label = np.array(label_data.get_data())
```

```python
for n in np.unique(label):
    plt.title(n)
    plt.imshow(np.where(label == n, 1, 0)[:,:, 24])
    plt.show()
```
